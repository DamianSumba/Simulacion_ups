## **Regresion Lineal**

La regresión lineal es un campo de estudio que enfatiza la relación estadística entre dos variables continuas conocidas como variables de predicción y respuesta . ... La variable predictora se denota con mayor frecuencia como x y también se conoce como variable independiente.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Regresión Lineal

def f(x): 
    np.random.seed(42) # 
    y = 0.1*x + 1.25 + 0.2*np.random.randn(x.shape[0]) #calculo de valores en base a una funcion
    return y

x = np.arange(0, 20, 0.5) 
y = f(x)

plt.scatter(x,y,label='data', color='green')
plt.grid(True)
plt.title('puntos generados');
```

```python
rl = LinearRegression() # Realizamos la regresion lineal de sklearn
rl.fit(x.reshape(-1,1), y) 
print("ECUACION DE LA RECTA F(X) = MX + B")
print('M = ' + str(rl.coef_) + ', b = ' + str(rl.intercept_))
```

```python
funcion_lineal=lambda x: rl.coef_*x+rl.intercept_
li=np.arange(min(x)-5.0,max(x)+5.0,0.5)
plt.scatter(x,y,label='data', color='red')
plt.plot(li,funcion_lineal(li))
plt.grid(True)
plt.show()
```

```python
nuevo_x = np.array([10]) 
prediccion = rl.predict(nuevo_x.reshape(-1,1))
print('nueva prediccion', prediccion)
```
